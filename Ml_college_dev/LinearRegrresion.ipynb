```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
print(diabetes.keys())

#converting data to Data frame
data = pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
data['target'] = diabetes.target
data.head()
```

```python
# splitting test and train file

# split the data into X and Y
X = data.drop('target',axis=1)
Y = data['target']
X.shape,Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train) # train the model on dataset
```

```python
# predict on testing set

Y_test_pred = model.predict(X_test)
Y_test_pred
```

```python
from sklearn.metrics import r2_score , mean_squared_error ,mean_absolute_error
r2 = r2_score(Y_test,Y_test_pred) #  varience of dependent variable
mse = mean_squared_error(Y_test,Y_test_pred)
mae = mean_absolute_error(Y_test,Y_test_pred)
intercept = model.intercept_
coef = model.coef_
print(r2,mse,mae,intercept,coef)
```

```python
import matplotlib.pyplot as plt
plt.scatter(Y_test,Y_test_pred,alpha = 0.5)
plt.plot(Y_test,Y_test,color='red')
plt.xlabel('Y_test')
plt.ylabel('Y_test_pred')
plt.title('Y_test vs Y_test_pred')
plt.show()
```
