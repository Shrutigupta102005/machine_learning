```python
'''                                               Data transformation
normalization  works with numeric data
 helps in computation(0-1 ka scale )
 Xnew = (X-Xmin)/(Xmax-Xmin)
'''
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
data = {
    "A":[30,40,50,60,70],
    "B":[10,20,30,40,50],
    "C":[100,200,300,400,500]
}

data1 = pd.DataFrame(data)
scaler = MinMaxScaler()

scaled_data = pd.DataFrame(scaler.fit_transform(data1),columns = data1.columns )# means all the  columns
print(scaled_data)

```

```python
'''
Standardization also known as z score normalisation
mean value = 0
standard deviation =1
xi = xi-mean(x)/std(x)
'''
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
data = {
    "A":[500,-500,300,-300,200,-300,120],
    "B":[10,20,30,40,50,60,70],
    "C":[100,200,300,400,500,600,700]
}

data1 = pd.DataFrame(data)
scaler = StandardScaler()

scaled_data = pd.DataFrame(scaler.fit_transform(data1),columns = data1.columns )# means all the  columns
print(scaled_data)

```
