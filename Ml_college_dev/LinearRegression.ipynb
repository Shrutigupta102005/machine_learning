```python
x = data[["YearsExperience"]]  # Use double brackets for x to keep it as a DataFrame
y = data[["Salary"]]           # Use double brackets to keep y as a DataFrame

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=20)

```

```python
len(x_train)
```

```python
model = LinearRegression()
```

```python
# training the model
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```
